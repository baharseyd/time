```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1/(1+np.exp(-x))
def sigmoid_derivative(x):
    return sigmoid(x) * (1-sigmoid(x))



data=pd.read_csv("C:\\file csv\\n.csv")
data



X=data.iloc[:, :6].values
y=data.iloc[:, 6].values
input_size=6
hidden_size1=4
hidden_size2=2
output_size=1
learning_rate=0.1
epochs=1000




np.random.seed(15)
weights_hidden1=np.random.uniform(size=(input_size, hidden_size1))
bias_hidden1=np.random.uniform(size=(1,hidden_size1))
weights_hidden2=np.random.uniform(size=(hidden_size1, hidden_size2))
bias_hidden2=np.random.uniform(size=(1, hidden_size2))
weights_output=np.random.uniform(size=(hidden_size2, output_size))
bias_output=np.random.uniform(size=(1, output_size))



for epoch in range(epochs):
    #freedforward
    hidden1_layer_activation=np.dot(X,weights_hidden1)+ bias_hidden1
    hidden1_layer_output=sigmoid(hidden1_layer_activation)
    hidden2_layer_activation=np.dot(hidden1_layer_output,weights_hidden2)+bias_hidden2
    hidden2_layer_output=sigmoid(hidden2_layer_activation)
    output_layer_activation=np.dot(hidden2_layer_output,weights_output)+bias_output
    predicted_output=sigmoid(output_layer_activation)
    
    
    #;
    error=y.reshape(-1,1) - predicted_output
    
    
    #Backpropagation
    d_predicted_output=error *sigmoid_derivative(output_layer_activation)
    error_hidden2=d_prediction_output.dot(weights_output.T)
    d_hidden2=error_hidden2 *sigmoid_derivative(hidden2_layer_activation)
    error_hidden1=d_hidden2.dot(weights_hidden2.T)
    d_hidden1=error_hidden1 * sigmoid_derivative(hidden1_layer_activation)
    
    weights_output+=  hidden2_layer_output.T.dot(d_predicted_output) * learning_rate
    bias_output+=np.sum(d_predicted_output,axis=0,keepdims=True)*learning_rate
    weights_hidden2+=hidden1_layer_output.T.dot(d_hidden2) * learning_rate
    bias_hidden2+=np.sum(d_hidden2,axis=0,keepdims=True)* learning_rate
    weights_hidden1+=X.T.dot(d_hidden1)* learning_rate
    bias_hidden1+=np.sum(d_hidden1,axis=0,keepdims=True)* learning_rate
    
    
#plot    
plt.plot(error)
plt.xlabel('Epochs')
plt.ylabel('Error')
plt.legend()
plt.show() 

    
    
counts,bins,patches=plt.hist(error,bins=5,orientation="horizontal",color="pink")
percentages=np.round(( counts/len(error))*100,2)
print(percentages)
for i in range(len(percentages)):
    plt.text(percentages[i],bins[i],str(percentages[i]), horizontalalignment="center",
    verticalalignment="center",fontsize=10,fontweight="bold")
plt.show() 

#mean error
losses=[]
for epoch in range(epochs):
    loss=0.5*((predicted_output - y)**2).mean()
    losses.append(loss)
np.mean(losses)
```
